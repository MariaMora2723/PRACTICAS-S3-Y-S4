# Group By

### 7. **¿Cuántos préstamos hay por cada estado y cual es la media de la cantidad de dinero prestado?**

```python
%%sql 
SELECT count(*)
FROM LOAN
```

Al aplicar el comando obtenemos que hay 682 préstamos. 

7.1 **¿CUÁNTOS PRESTAMOS HAY POR CADA ESTADO?**

```python
%%sql
SELECT status, count (*) as total
FROM LOAN
GROUP BY status
ORDER BY total DESC
```

Con el comando anterior me ha seleccionado los préstamos por tipos de estados (con el as total le indico que me nombre esa columna como total), me lo indica por grupos de estado y me ordena el total de forma descendiente. 


7.2 **¿Cual es la media de la cantidad de dinero prestado?**

```python
%%sql
SELECT status, count (*) as total, avg (amount)
FROM LOAN
GROUP BY status
ORDER BY total DESC
```

La última línea AVG es la que me indica la media de cada estado. 

7.3  ***Sácame cuáles son los que tienen una duración igual o mayor a 24. ***

```python
%%sql
SELECT status, count (*) as total, avg (amount)
FROM LOAN
WHERE duration >= 24
GROUP BY status
ORDER BY total DESC
```

La última línea AVG es la que me indica la media de cada estado. 

7.4 **Ahora solo quiero que me muestre los grupos que tienen mas de 25 prestamos concedidos.** 

where - se filtran los registros antes de agruparlos
habing - se filtran los grupos tras obtenerlos

```python
%%sql
SELECT status, count (*) as total, avg (amount)
FROM LOAN
WHERE duration >= 24
GROUP BY status
HAVING total > 25
ORDER BY total DESC
```

La última línea AVG es la que me indica la media de cada estado.

8.1 **Ahora sacamos todos los campos que hay dentro de esas tablas, es decir, todo lo que las componen.** 

```python
%%sql
SELECT loan_id, l.date loan_date, amount, duration, payments, frequency, a.date acc_date
FROM LOAN l, ACCOUNT a
WHERE l.account_id = a.account_id
AND status in ("A", "B")
LIMIT 10
```

Como la columna account la tengo tanto en la tabla LOAN como en ACCOUNT, se lo indico poniendoles un igual. 

8.3 **Úneme las tablas de "ACCOUNT" y "LOAN"**

```python
%%sql
SELECT *
FROM LOAN l, ACCOUNT a
WHERE l.account_id = a.account_id
LIMIT 50
```

8.4 **¿Cómo hago para vincular también la tabla de distric a esta?**

```python
%%sql
SELECT loan_id, A4, A9, A10, A11, l.date loan_date, amount, duration, payments, frequency, a.date acc_date
FROM LOAN l, ACCOUNT a, DISTRICT d
WHERE l.account_id = a.account_id
AND a.district_id = d.A1
AND status in ("A", "B")
ORDER BY amount DESC, a.date DESC
LIMIT 10
```

8.5 **Quiero añadir el año de nacimiento del cliente a estos datos.**

```python
%%sql
SELECT loan_id, A4, A9, A10, A11,c.birth_number, l.date loan_date, amount, duration, payments, frequency, a.date acc_date
FROM LOAN l, ACCOUNT a, DISTRICT d, DISPOSITION dp, CLIENT c
WHERE l.account_id = a.account_id
AND a.district_id = d.A1
AND dp.account_id = a.account_id -- join disposition y account
AND dp.client_id = c.client_id  -- join disposition y client
AND status in ("A", "B")
ORDER BY amount DESC, a.date DESC
LIMIT 10
```

Con disposition hacemos una tabla puente, ya que es la única con la que tanto cliente como account tienen las dos relación, para poder unir account con client y con eso poder relacionar el cumpleaños con la cuenta de los clientes. 

8.6 ***Ahora vamos a filtrar por el tipo. ***

```python
%%sql
SELECT count(*) --loan_id, A4, A9, A10, A11,c.birth_number, l.date loan_date, amount, duration, payments, frequency, a.date acc_date
FROM LOAN l, ACCOUNT a, DISTRICT d, DISPOSITION dp, CLIENT c
WHERE l.account_id = a.account_id
AND a.district_id = d.A1
AND dp.account_id = a.account_id -- join disposition y account
AND dp.client_id = c.client_id  -- join disposition y client
AND dp.type = "OWNER"
--AND status in ("A", "B")
--ORDER BY amount DESC, a.date DESC
--LIMIT 10
```

Esto nos ha indicado que hay 682 propietarios de cuentas. 

8.7 **Ahora vamos a sacar el año de nacimiento**

```python
%%sql
SELECT loan_id, A4, A9, A10, A11,c.birth_number, SUBSTR (c.birth_number, 0,3),l.date loan_date, amount, duration, payments, frequency, a.date acc_date
FROM LOAN l, ACCOUNT a, DISTRICT d, DISPOSITION dp, CLIENT c
WHERE l.account_id = a.account_id
AND a.district_id = d.A1
AND dp.account_id = a.account_id -- join disposition y account
AND dp.client_id = c.client_id  -- join disposition y client
AND dp.type = "OWNER"
AND status in ("A", "B")
ORDER BY amount DESC, a.date DESC
LIMIT 10
```

La función SUBSTR nos modifica un texto, en este caso le meto ese comando a la fecha de cumpleaños para que me coja los 2 primeros digitos (le indico de 0 a 3 porque al ser un intervalo abierto no me coge las puntas).

En el caso de que estuviera dividido como numero entero tendríamos que dividirlo entre 1000.

```python
%%sql
SELECT c.client_id, avg (amount) as av
from client c, district d, disposition ds, account a, trans t
WHERE c.district_id = d.A1 and c.client_id = ds.client_id
    and ds.account_id = a.account_id and a.account_id = t.account_id
and d.A2 = "Kolin" and ds.type = "OWNER" 
GROUP BY c.client_id
ORDER BY av DESC
LIMIT 10

```
